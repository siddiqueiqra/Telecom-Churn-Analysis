### Summary tables

```python
#building contingency table 
pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

```python
# most of the users are loyal and do not use additional services(International Plan/Voice mail)
```

### Pivot tables

```python
#take a look at the average number of day, evening, and night calls by area code
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```

### DataFrame Transformation

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
```

```python
df.head()
```

```python
df.drop(['Total call'], axis=1)
```

```python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

### Predicting Telecom Churn

```python
#seeing how churn rate is related to the International plan feature.
pd.crosstab(df["Churn"], df["International plan"], margins=True)
```
